```python
import requests
from bs4 import BeautifulSoup
import csv
import time
import random

# URL of the website
fortunes = []

for idx in range(1, 16):
    print(f"Going through iteration {idx}")
    url = f'http://www.fortunecookiemessage.com/archive.php?start={50*idx}'  # Replace with the actual URL

    # Send a GET request to the website
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    counter = 0

    # Find all the <tr> tags, and then extract the text from the <a> tags within
    for tr in soup.find_all('tr'):
        a_tag = tr.find('a')
        if a_tag and 'href' in a_tag.attrs and '/cookie/' in a_tag['href']:
            counter += 1
            print(f"Currently processing Number {counter}/50")
            fortunes.append(a_tag.get_text(strip=True))

            # Sleep for a random duration between requests to mimic human behavior
            time.sleep(random.uniform(0.5, 3))  # Waits for 0.5 to 2 seconds
```

```python
import pandas as pd

fortune_df = pd.DataFrame({"fortunes": fortunes})
fortune_df.head()
```

```python
fortune_df.to_csv("data/fortunecookiemessage_com_archive.csv", index=False)
```
